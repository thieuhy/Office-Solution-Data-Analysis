```python
CustomerNameShipping = SalesData[SalesData["Ship Mode"] == "Same Day"][["Customer Name", "Ship Mode"]]
display(CustomerNameShipping.head(10))
```

```python
repeat_customers = SalesData[SalesData["Ship Mode"] == "Same Day"]["Customer Name"].value_counts()
repeat_customers = repeat_customers[repeat_customers > 1]  # Only customers with multiple orders
print("Number of Repeat Same-Day Customers:" + str(len(repeat_customers)))
repeat_customers.head(10)
```

```python
import matplotlib.pyplot as plt

repeat_customers.head(10).plot(kind="bar", figsize=(10,5), color="blue")
plt.xlabel("Customer Name")
plt.ylabel("Number of Same-Day Orders")
plt.title("Top 10 Repeat Customers for Same-Day Shipping")
plt.xticks(rotation=45)
plt.show()
```

```python
SalesData["Order Date"] = pd.to_datetime(SalesData["Order Date"])

# Filter Same-Day orders from repeat customers
repeat_orders = SalesData[(SalesData["Ship Mode"] == "Same Day") &
                          (SalesData["Customer Name"].isin(repeat_customers.index))]

# Group by month
repeat_orders_over_time = repeat_orders.groupby(repeat_orders["Order Date"].dt.to_period("M"))["Customer Name"].count()

# Plot trend
plt.figure(figsize=(12,6))
repeat_orders_over_time.plot(kind="line", marker="o", color="green")
plt.xlabel("Month")
plt.ylabel("Number of Repeat Same-Day Orders")
plt.title("Trend of Repeat Customers Using Same-Day Shipping")
plt.grid()
plt.show()
```

```python
# Filter Same-Day orders
same_day_orders = SalesData[SalesData["Ship Mode"] == "Same Day"]

# Identify customers who used to order Same-Day but stopped in the most recent year
last_year = SalesData["Order Date"].dt.year.max()
previous_customers = same_day_orders[same_day_orders["Order Date"].dt.year < last_year]["Customer Name"].unique()
current_customers = same_day_orders[same_day_orders["Order Date"].dt.year == last_year]["Customer Name"].unique()

# Find lost customers
lost_customers = set(previous_customers) - set(current_customers)

# Print lost customers
print("Number of Lost Same-Day Customers:", len(lost_customers))
print("Lost Customers:", list(lost_customers)[:10])  # Display first 10 lost customers
```

```python
# Compare average sales price & discount over time
# Create a 'Year-Month' column from 'Order Date'
same_day_orders.loc[:, "Year-Month"] = same_day_orders["Order Date"].dt.to_period("M")  # Use .loc to avoid the warning

same_day_price_discount = same_day_orders.groupby("Year-Month")[["Sales", "Discount"]].mean()

# Get recent vs previous average sales price & discount
latest_avg_sales = same_day_price_discount.iloc[-1]["Sales"]
previous_avg_sales = same_day_price_discount.iloc[0]["Sales"]
latest_avg_discount = same_day_price_discount.iloc[-1]["Discount"]
previous_avg_discount = same_day_price_discount.iloc[0]["Discount"]

print("Latest Avg Sale Price: ${:.2f}, Previous Avg Sale Price: ${:.2f}".format(latest_avg_sales, previous_avg_sales))
print("Latest Avg Discount: {:.2%}, Previous Avg Discount: {:.2%}".format(latest_avg_discount, previous_avg_discount))
```

```python
# Identify regions with largest drops in Same-Day orders
region_orders = same_day_orders.groupby(["Year-Month", "Region"])["Order ID"].count().unstack()

# Compare most recent and previous year regional data
print("Same-Day Orders by Region:")
print(region_orders.tail(12))  # Show last 12 months
```

```python
# Get top 10 most expensive Same-Day orders
high_cost_orders = same_day_orders.sort_values(by="Sales", ascending=False).head(10)[["Order ID", "Customer Name", "Sales", "Profit", "Region", "Segment"]]
print("Top 10 High-Cost Same-Day Orders:")
print(high_cost_orders)
```

```python
import matplotlib.pyplot as plt

# Count same-day orders per region
same_day_by_region = SalesData[SalesData["Ship Mode"] == "Same Day"]["Region"].value_counts()

# Plot top regions for same-day shipping
same_day_by_region.head(10).plot(kind="bar", figsize=(10,5), color="blue")
plt.xlabel("Region")
plt.ylabel("Number of Same-Day Orders")
plt.title("Top Regions for Same-Day Shipping")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count same-day orders per postal code
same_day_by_postal = SalesData[SalesData["Ship Mode"] == "Same Day"]["Postal Code"].value_counts()

# Show top 10 postal codes
same_day_by_postal.head(10)
```

```python
# Create profitability_by_shipping

profitability_by_shipping = SalesData.groupby("Ship Mode")[["Sales", "Profit"]].sum()
profitability_by_shipping["Profit Margin (%)"] = (profitability_by_shipping["Profit"] / profitability_by_shipping["Sales"]) * 100

# Formating in $
profitability_by_shipping["Sales"] = profitability_by_shipping["Sales"].map("${:,.2f}".format)
profitability_by_shipping["Profit"] = profitability_by_shipping["Profit"].map("${:,.2f}".format)
profitability_by_shipping["Profit Margin (%)"] = profitability_by_shipping["Profit Margin (%)"].map("{:.2f}%".format)

# Print the formatted table
print(profitability_by_shipping)
```

```python
# Calculate average order value per shipping mode
avg_order_value = SalesData.groupby("Ship Mode")["Sales"].mean().sort_values()

print("Average Order Value per Shipping Mode:\n", avg_order_value)
```

```python
# Count product categories ordered via Same-Day
same_day_products = SalesData[SalesData["Ship Mode"] == "Same Day"]["Category"].value_counts()

print("Top Product Categories for Same-Day Orders:\n", same_day_products)
```

```python
# Check average discount by shipping mode
avg_discount = SalesData.groupby("Ship Mode")["Discount"].mean()

print("Average Discount per Shipping Mode:\n", avg_discount)
```

```python
# Compare product category choices for repeat vs. one-time Same-Day users
repeat_customers = SalesData[SalesData["Ship Mode"] == "Same Day"]["Customer Name"].value_counts()
repeat_customers = repeat_customers[repeat_customers > 1].index

# Get product choices of repeat customers
repeat_same_day_products = SalesData[(SalesData["Ship Mode"] == "Same Day") &
                                     (SalesData["Customer Name"].isin(repeat_customers))]["Category"].value_counts()

print("Top Product Categories for Repeat Same-Day Customers:\n", repeat_same_day_products)
```

```python
# Compare Same-Day Shipping trends to other modes
same_day_trend = SalesData[SalesData["Ship Mode"] == "Same Day"].groupby(SalesData["Order Date"].dt.to_period("M"))["Sales"].sum()
standard_trend = SalesData[SalesData["Ship Mode"] == "Standard Class"].groupby(SalesData["Order Date"].dt.to_period("M"))["Sales"].sum()

# Plot trends
plt.figure(figsize=(12,6))
same_day_trend.plot(kind="line", marker="o", color="red", label="Same-Day Sales")
standard_trend.plot(kind="line", marker="o", color="blue", label="Standard Class Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.title("Sales Trend: Same-Day vs Standard Class")
plt.legend()
plt.grid()
plt.show()
```
